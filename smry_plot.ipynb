```python
import matplotlib.pyplot as plt
 
## 데이터
age_category = ['10대 이하', '20대', '30대', '40대', '50대', '60대 이상'] ## 연령 카테고리, x축 눈금에 표시될 라벨
num_patient = [3,3,5,7,8,9] ## 방문환자수
 
## 시각화
plt.figure(figsize=(5,5)) ## Figure 생성 사이즈는 10 by 10


# X 축 위치, 라벨 설정 
xtick_label_position = list(range(len(age_category))) ## x축 눈금 라벨이 표시될 x좌표
plt.xticks(xtick_label_position, age_category) ## x축 눈금 라벨 출력 

plt.bar(xtick_label_position, num_patient) ## 바차트 출력

plt.title('지난달 연령별 방문환자 수',fontsize=20) ## 타이틀 출력
plt.xlabel('연령') ## x축 라벨 출력
plt.ylabel('방문환자수') ## y축 라벨 출력
plt.show()
```

```python
df = sns.load_dataset('titanic')
temp = df.notna().sum()

plt.figure(figsize=(12, 6));
plt.bar(x=temp.index, height=temp);


plt.title('TITLE', fontsize=18);
plt.ylabel('Counts');

plt.xticks(rotation=45); # 글자 45도 
plt.show();
```

```python
temp = df.groupby(['sex', 'survived'])['fare'].mean().reset_index()


# 시각화 
fig, ax = plt.subplots(figsize=(8, 6)) # 크기
ax.set_title("TITLE")   # 제목

sns.barplot(data=temp, x='survived', y='fare', hue='sex', ax=ax)

ax.set_xticklabels(['Not Survived', 'Survived' ])    #  X축 표시 값  #(!) 그래프 그린 이후에 추가 해 주어야 함

plt.show()
```

# 히스토그램

```python
fig, ax = plt.subplots(figsize=(8, 6))

bins = np.arange(0, df.age.max()+df.age.max()/8, df.age.max()/8)
fig = plt.hist(x=df['age'], bins=bins)

plt.title('AGE dist', fontsize=16)  # 제목
plt.xlabel('age')   # X축 설명
plt.ylabel('count') # y축 설명


plt.show()
```

```python
# Seabor count plot

fig, ax = plt.subplots(figsize=(8, 6)
    , facecolor = 'yellow' # 배경색
    , alpha=1 #(TODO) 투명도 인데 이곳에서 셋팅하면 영향을 안받음
    )
#fig.set_facecolor = 'yellow'  # 안됨   
fig.patch.set_alpha( 0.2) # 배경색 투명도

ax.set_facecolor ( 'blue') # 배경색
ax.patch.set_alpha( 0.1 ) #(TODO) 투명도 인데 영향을 모르겠음 

ax.tick_params(axis='x', colors='black') # x 축 눈금 색깔
ax.tick_params(axis='y', colors='black') # y 축 눈금 색깔



# 그래프 
sns.countplot(x=df['pclass'].values, ax=ax)

# 부분 강조 박스 case 01
ax.axvspan(1,3, facecolor='gray', alpha=0.6) # x가 1~3에 해당하는 부분 배경색 설정
 
# 부분 강조 박스 case 02
## (1, 100)에서 시작하여 가로 3, 세로 50 인 빨간색 사각형
ax.add_patch(
     patches.Rectangle(
        (1, 100),
        3,
        50,
        facecolor = 'red',
        fill=True,
        alpha=0.5
     ) ) 

# x축, y축 최대 표현 범위
ax.set_xlim(-2, 10)
ax.set_ylim(-1, 800)


plt.title('count by class', fontsize=16)
plt.show()
```

```python


# 가상 데이터 
x_col_name = list(range(0, 10))
y_col_name_1 = np.random.normal(10, 2, size=(10))
y_col_name_2 = np.random.normal(10, 3, size=(10))
df_tmp = pd.DataFrame({'y_col_name_1': y_col_name_1, 
                       'y_col_name_2': y_col_name_2})
df_tmp = df_tmp.reset_index().rename(columns={'index': 'x_col_name'})


text_font_size = 10

fig, ax = plt.subplots(figsize=(8, 8))

x_col_name = 'x_col_name'

# 계산 시간 막대 그래프
y_col_name = "y_col_name_1"
ax1 = sns.barplot ( data=df_tmp , x=df_tmp[x_col_name]  , y=df_tmp[y_col_name] , label=y_col_name
    , color='b' , alpha=0.2 
    , ax=ax
) #, ax=ax1)
# plt.ylim ( 0, 100 )

# 성능지표 선 그래프
ax2 = ax1.twinx()
y_col_name = "y_col_name_1"
ax2 = sns.lineplot ( data=df_tmp , x=df_tmp[x_col_name] , y=df_tmp[y_col_name] , label=y_col_name
    , alpha=0.2 , marker='o' #, color='b'#, sort = False 
    , ax=ax2)

y_col_name = "y_col_name_2"
ax2 = sns.lineplot ( data=df_tmp , x=df_tmp[x_col_name] , y=df_tmp[y_col_name] , label=y_col_name
    , alpha=0.2 , marker='o' #, color='b'#, sort = False 
    , ax=ax2)


# 숫자 넣는 부분, height + 0.25로 숫자 약간 위로 위치하게 조정
y_col_name = "y_col_name_1"
for x , y in zip( df_tmp[x_col_name] , df_tmp[y_col_name] ) :
    plt.text(x, y,   round( y, 4)
    , ha='center', va='bottom', size = text_font_size, color = 'black')

# plt.ylim ( 0.5, 0.9 ) 

#(v)) 선 그래프와 막대 그래프의 X축 위치가 안 맞는 문제
#(v) 시본에서 막대그래프 + 선 그래프 출력시 라인그래프 밀리는 증상 수정
ax2.set(xticks=range(10), xticklabels=list(range(1,11)))

title = "성능지표 vs 시간" 
ax.set_title(title)
```
